## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### 💡 Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
"""
My Budget Tracker Prompt:
[Create a Python program that allows users to input their monthly income, input up to 3 expenses, calculates remaining budget and shows a warning if budget is negative ]
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# Monthly Budget Calculator

# Input monthly income
income = float(input("Enter your monthly income: "))

# Input up to 3 expenses
expenses = []
for i in range(1, 4):
    expense = input(f"Enter expense {i} (or press Enter to skip): ")

    if expense == "":
        break

    expenses.append(float(expense))

# Calculate total expenses
total_expenses = sum(expenses)

# Calculate remaining budget
remaining_budget = income - total_expenses

# Display results
print("\n----- Budget Summary -----")
print(f"Total Income: ${income:.2f}")
print(f"Total Expenses: ${total_expenses:.2f}")
print(f"Remaining Budget: ${remaining_budget:.2f}")

# Warning if negative
if remaining_budget < 0:
    print("⚠️ WARNING: Your budget is negative! You are overspending.")
else:
    print("✅ Good job! Your budget is within limits.")
```

```python
# REFINED VERSION - Your Working Budget Tracker
from datetime import datetime

print("=== Monthly Budget Tracker ===")

timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

# Get user income
while True:
    try:
        income = float(input("Enter your monthly income: $"))
        if income < 0:
            print("Income cannot be negative. Try again.")
            continue
        break
    except ValueError:
        print("Please enter a valid number.")

# Get expenses
expenses = []
for i in range(3):
    while True:
        try:
            expense = float(input(f"Enter expense {i+1}: $"))
            if expense < 0:
                print("Expense cannot be negative. Try again.")
                continue
            expenses.append(expense)
            break
        except ValueError:
            print("Please enter a valid number.")

# Calculate totals
total_expenses = sum(expenses)
balance = income - total_expenses

# Display results
print("\n===== Budget Summary =====")
print(f"Date: {timestamp}")
print(f"Total Income: ${income:.2f}")
print(f"Total Expenses: ${total_expenses:.2f}")
print(f"Remaining Balance: ${balance:.2f}")

if balance > 0:
    print("Great job! You are saving money 👍")
elif balance == 0:
    print("You broke even this month.")
else:
    print("Warning: You are overspending ⚠️")
```

## Part 4: Enhancement Challenge

Choose ONE feature to add using AI assistance:
- 🏷️ Expense categories (food, transport, entertainment)
- 📊 Show percentage of income spent
- ✅ Input validation (no negative numbers)
- 📝 Detailed budget summary with advice

```python
# Document your enhancement
enhancement_choice = "I chose to add: expense categories"
enhancement_prompt = """
[Modify my budget tracker so that each expense asks for a category
(food, transport, entertainment)]
"""

print(f"🚀 ENHANCEMENT: {enhancement_choice}")
```

```python
# ENHANCED BUDGET TRACKER
from datetime import datetime

print("=== Enhanced Monthly Budget Tracker ===")

timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

# Get income
while True:
    try:
        income = float(input("Enter your monthly income: $"))
        if income < 0:
            print("Income cannot be negative.")
            continue
        break
    except ValueError:
        print("Please enter a valid number.")

# Categories dictionary
categories = {
    "food": 0,
    "transport": 0,
    "entertainment": 0
}

# Get expenses
for i in range(3):
    while True:
        try:
            amount = float(input(f"Enter expense amount {i+1}: $"))
            if amount < 0:
                print("Expense cannot be negative.")
                continue
            break
        except ValueError:
            print("Please enter a valid number.")

    while True:
        category = input("Enter category (food/transport/entertainment): ").lower()
        if category in categories:
            categories[category] += amount
            break
        else:
            print("Invalid category. Choose food, transport, or entertainment.")

# Calculations
total_expenses = sum(categories.values())
balance = income - total_expenses

# Display results
print("\n===== Detailed Budget Summary =====")
print(f"Date: {timestamp}")
print(f"Income: ${income:.2f}")
print("\nExpenses by Category:")
for cat, value in categories.items():
    print(f"{cat.capitalize()}: ${value:.2f}")

print(f"\nTotal Expenses: ${total_expenses:.2f}")
print(f"Remaining Balance: ${balance:.2f}")

# Advice
if balance > 0:
    print("Good job! You are managing your budget well 👍")
elif balance == 0:
    print("You spent exactly your income.")
else:
    print("Warning: You are overspending ⚠️")
```
